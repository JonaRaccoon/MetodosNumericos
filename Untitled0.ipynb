```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)

y = x**2
plt.plot(x, y)


plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de y = x^2")


plt.grid(False)
plt.show()
```
