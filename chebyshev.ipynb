```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

def chebyshev(n, a, b):
    k = np.arange(n)
    x = np.cos((2 * k + 1) * np.pi / (2 * n))
    espaciados_x = 0.5 * (b - a) * x + 0.5 * (a + b)
    return espaciados_x

def runge(x):
    return 1 / (1 + 25 * x**2)

a, b = -1, 1

num_nodos = [5, 9, 21]

x_plot = np.linspace(a, b, 200)
plt.figure(figsize=(10, 6))
plt.plot(x_plot, runge(x_plot), label="Runge", color='black')

for n in num_nodos:
    nodos = chebyshev(n, a, b)
    values = runge(nodos)
    poly = lagrange(nodos, values)
    y_interp = poly(x_plot)

    plt.plot(x_plot, y_interp, label=f'Lagrange ({n} nodos)')
    plt.plot(nodos, values, 'o', label=f'Nodos ({n})' if n == num_nodos[-1] else "")

plt.title("Interpolacion de Lagrange")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
